```python
import cv2
import os
import math
from ultralytics import YOLO
import winsound
import time
from collections import defaultdict
import numpy as np

# Load the YOLO models (using Bagging)
model_paths = [
    'C:\\Users\\HP\\Desktop\\fall\\runs\\detect\\train20\\weights\\best.pt'  # Add more models as needed
]

models = [YOLO(model_path) for model_path in model_paths]

# Load class names
classnames = []
with open('C:\\Users\\HP\\Desktop\\fall\\classnames.txt', 'r') as f:
    classnames = f.read().splitlines()

# Create output directory
output_folder = 'C:\\Users\\HP\\Desktop\\fall\\output_human'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

fall_confidences = []
frame_count_since_last_check = 0

def calculate_distance(p1, p2):
    return math.sqrt((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2)

def calculate_angle(a, b, c):
    return math.degrees(math.acos((a**2 + b**2 - c**2) / (2 * a * b)))

def process_frame(frame, confidence_threshold=0.5):
    combined_results = defaultdict(list)

    for model in models:
        results = model(frame)
        for result in results:
            for box in result.boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])

                confidence = box.conf.cpu().numpy()[0]  # Extract as a scalar
                class_detect = int(box.cls[0].cpu().numpy())  # Move to CPU and convert to NumPy
                class_detect = classnames[class_detect]
                
                combined_results[class_detect].append(confidence)

    # Aggregate predictions (Bagging approach)
    for class_detect, confidences in combined_results.items():
        avg_confidence = np.mean(confidences)  # Average confidence

        if avg_confidence > confidence_threshold and class_detect == 'person':
            # Using the bounding box from the first model for visualization
            x1, y1, x2, y2 = map(int, result.boxes[0].xyxy[0])  # Use the bounding box of the first detected object

            # Calculate dimensions and centroid
            centroid = ((x1 + x2) / 2, (y1 + y2) / 2)
            upper_left = (x1, y1)
            upper_right = (x2, y1)
            height = y2 - y1
            width = x2 - x1

            # Human fall detection
            threshold = height - width
            a = calculate_distance(upper_left, upper_right)  # Base
            b = calculate_distance(upper_right, centroid)     # Side 2
            c = calculate_distance(upper_left, centroid)      # Side 3

            # Calculate angles
            angle_C = 180 - (calculate_angle(a, b, c) * 2)  # Angle at centroid

            if 0 < angle_C:
                # Check fall condition based on angle
                if (angle_C > 90 or angle_C < 10) or threshold < 0:
                    color = (0, 0, 255)  # Red for fall detection
                    label = f"Fall Detected ({avg_confidence:.2f})"  # Use avg_confidence
                    fall_confidences.append(avg_confidence)  # Store the average confidence value

                    # Draw the bounding box and label
                    cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
                    cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    return frame

def continuous_beep(duration):
    end_time = time.time() + duration
    while time.time() < end_time:
        winsound.Beep(1500, 200)
        time.sleep(0.2)

def process_input(input_source, confidence_threshold=0.25):
    global frame_count_since_last_check
    frame_count = 0
    output_file_path = os.path.join(output_folder, os.path.basename(input_source))

    if input_source.endswith(('.mp4', '.avi', '.mov', '.gif', '.webm')):
        cap = cv2.VideoCapture(input_source)
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        fps = cap.get(cv2.CAP_PROP_FPS)

        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        output_video = cv2.VideoWriter(output_file_path, fourcc, fps, (width, height))

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            processed_frame = process_frame(frame, confidence_threshold)
            output_video.write(processed_frame)
            cv2.imshow('Detections', processed_frame)

            frame_count += 1
            frame_count_since_last_check += 1
            
            if frame_count_since_last_check >= fps * 3:
                if fall_confidences:
                    average_confidence = sum(fall_confidences) / len(fall_confidences)
                    print(f"Average confidence over last 5 seconds: {average_confidence:.2f}")
                    if average_confidence > 0.30:
                        continuous_beep(3)

                fall_confidences.clear()
                frame_count_since_last_check = 0

            if cv2.waitKey(50) & 0xFF == ord('q'):
                break

        cap.release()
        output_video.release()

    else:
        frame = cv2.imread(input_source)
        processed_frame = process_frame(frame, confidence_threshold)
        output_image_path = os.path.join(output_folder, os.path.basename(input_source))
        cv2.imwrite(output_image_path, processed_frame)
        cv2.imshow('Detections', processed_frame)
        cv2.waitKey(0)

    cv2.destroyAllWindows()

# Example usage
input_path = 'C:\\Users\\HP\\Desktop\\fall\\T\\Fall 2\\Human fall skatting .mp4'
process_input(input_path, confidence_threshold=0.25)
```
